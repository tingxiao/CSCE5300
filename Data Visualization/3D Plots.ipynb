## 1. Contour Plot

Letâ€™s make a contour plot! This code will make random data and create filled and unfilled contour plots! It will also label the contour plots and make various color mappings, which can be changed through setting a different cmap value. 

Here is a link that will be especially useful for your [homework assignment](https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html).  

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
#Following line is for color mapping
from matplotlib import cm
```



* This link shows the [various color maps](https://matplotlib.org/examples/color/colormaps_reference.html) that can be implemented into contour plots; their values can be used as the cmap value in the <a href= "https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html"> contour()</a> function. 

* This link describes how to [create different styles](https://www.python-course.eu/matplotlib_contour_plot.php) of contour plots, such as a meshgrid and a filled contour. 

* We can use pyplot's <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.clabel.html"> clabel() </a> function to add labels to a contour plot. 

<b>In our first example, an unfilled contour is created; in the second and third example, a filled contour is created. </b>

```python
#Creates an unfilled contour with a cool & warm color mapping

#Creating the data for the contour plot(s)
delta = 0.025
x = np.arange(-2.0,3.0,delta)
y = np.arange(-2.0,3.0,delta)
#meshgrid makes rectangular arrays that
#cover every combination of x and y values 
X, Y = np.meshgrid(x,y)

#make an interesting bump and valley for plotting
#Creates a Gaussian distribution for equal shape X, Y
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

#Creating an unfilled contour plot
plt.figure()
#Can change the color mappings and line style
cp = plt.contour(X, Y, Z,cmap=cm.coolwarm)

#Labels and titles for the plot
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

* To create a legend for the colors on a contour plot, we can use pyplot's <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.colorbar.html"> colorbar() </a> function. 
* The difference between generating a filled and unfilled contour plot is using either pyplot's <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contourf.html"> contourf() </a> or <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html"> contour() </a> functions. 


```python
#Creates a filled contour with a cool & warm color mapping

#Creating the data for the contour plot(s)
delta = 0.025
x = np.arange(-2.0,3.0,delta)
y = np.arange(-2.0,3.0,delta)
#meshgrid makes rectangular arrays that
#cover every combination of x and y values 
X, Y = np.meshgrid(x,y)

Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

#print(Z) -- Use this statement if you want to get the Z-values
#Creating a filled contour plot
plt.figure()
#Creating the contour plot with the data values and creating labels
#Filling the contour plot with designated colors
contour_filled = plt.contourf(X,Y,Z,cmap=cm.coolwarm)

#Creating a legend, title, and labels 
plt.clabel(contour_filled, colors='k',fmt = '%2.1f',fontsize=12)
plt.colorbar(contour_filled)
plt.title('Filled Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

```python
#Creates a filled contour with a black & white color mapping

#Creating the data for the contour plot(s)
delta = 0.025
x = np.arange(-2.0,3.0,delta)
y = np.arange(-2.0,3.0,delta)
#meshgrid makes rectangular arrays that
#cover every combination of x and y values 
X, Y = np.meshgrid(x,y)

Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

#print(Z) -- Use this statement if you want to get the Z-values
#Creating a filled contour plot
plt.figure()
#Creating the contour plot with the data values and creating labels
cp = plt.contourf(X, Y, Z)
plt.clabel(cp, colors='k',fmt = '%2.1f',fontsize=12)
#Filling the contour plot with a black and white color mapping
contour_filled = plt.contourf(X,Y,Z,cmap=cm.binary)

#Creating a legend, title, and labels 
plt.colorbar(contour_filled)
plt.title('Filled Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

## 2. Surface Plot

Letâ€™s make a surface plot! This code will formulate random data to create a surface plot with various color mappings. It also creates x, y, and z-tick marks, a color bar, and labels for the axes and the title. With this code, you can also change the angle of the surface plot to get views from every angle!

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

* This link creates [a surface plot with a cool & warm color mapping]( https://matplotlib.org/examples/mplot3d/surface3d_demo.html). 

* This link describes how to [create a colorbar]( https://stackoverflow.com/questions/18195758/set-matplotlib-colorbar-size-to-match-graph), so that, it matches with the size of the figure.

* This link describes how to [create labels]( https://matplotlib.org/examples/mplot3d/text3d_demo.html) in the x-, y-, and z-dimensions, since the surface plot is in the 3D dimension. 

* This link also explains the [ticker module]( https://matplotlib.org/api/ticker_api.html) and the classes that are contained within it; in our example, we use the LinearLocator class and the FormatStrFormatter class for tick formatting. 

```python
#Demonstrates plotting a 3D surface colored with the cool-warm color map.
#Surface plot is viewed at a 45-degree angle

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Setting rotation angle of plot
ax.view_init(azim=45) 

# Make data.
X = np.arange(-5, 5, 0.25) #array of evenly spaced values
Y = np.arange(-5, 5, 0.25) #array of evenly spaced values
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) #Creating the Z curve
# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```

* This link explains how to use the view_init function to [manipulate the angle]( https://stackoverflow.com/questions/33084853/set-matplotlib-view-to-be-normal-to-the-x-y-plane-in-python) of the surface plot. 

```python
#Demonstrates plotting a 3D surface colored with the cool-warm color map.
#Viewed from a 90-degree angle

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Setting rotation angle of plot
ax.view_init(azim=90) 

# Make data.
X = np.arange(-5, 5, 0.25) #array of evenly spaced values
Y = np.arange(-5, 5, 0.25) #array of evenly spaced values
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) #Creating the Z curve
# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```

## 3. Mesh Plot

Letâ€™s make a mesh plot! This code will make random data and create a wireframe plot (as itâ€™s called in Python). You can also change the color mapping of the plot and the angle itâ€™s viewed from! This code also allows you to label the axes and create a title. 

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

* This link describes how to [create a colored mesh plot]( https://stackoverflow.com/questions/28751535/colored-mesh-plot-of-a-function-in-python); in our example, we use data from the [axes3d object](https://matplotlib.org/2.0.2/mpl_toolkits/mplot3d/api.html) to plot our mesh plot. The following link is the Axes3D API. 

* This link explains how to [create the mesh figure in Python]( https://stackoverflow.com/questions/24909256/how-to-obtain-3d-colored-surface-via-python/24958192), since it's slightly more complicated than creating a mesh plot in Octave or Matlab. 

* This link describes how to [color the mesh plot]( https://stackoverflow.com/questions/15134004/colored-wireframe-plot-in-matplotlib) according to the plots' z-value; in other words, it explains how to get the colors from the color mapping on the lines of the mesh plot. 


```python
#Known as a wireframe plot in python
#Creates a wireframe plot with a plasma color mapping-- viewed at a 45 degree angle 

# Grab some test data.
#X,Y,Z are data values that are created as 2D arrays
X, Y, Z = axes3d.get_test_data(0.3) 

#Normalize to [0,1]-- used for color mapping
Z = (Z-Z.min())/(Z.max()-Z.min())

#Color mapping
colors = cm.plasma(Z) #Can customize with various color mapping styles,rcount,ccount,_ = colors.shape

# Plot a basic wireframe.
#rstride = arrayrowstride(step size)
#cstrode = arraycolumnstride(step size)
fig = plt.figure()
ax = fig.gca(projection='3d')

#Setting rotation angle of plot to 45 degrees 
ax.view_init(azim=45)

#rccount= max amount of rows
#ccount = max amount of columns 

surf= ax.plot_surface(X, Y, Z, rcount=100, ccount=100,facecolors=colors, shade=False)

#Setting the tick locators-- choosing tick locations

#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

#Add labels for x,y, and z axes
ax.set_xlabel('X axis', fontsize=10,labelpad=4.2)
ax.set_ylabel('Y axis', fontsize = 10,labelpad=4.2)
ax.set_zlabel('Z axis',fontsize=10,labelpad = 4.2)
ax.set_title('Mesh Plot')
surf.set_facecolor((0,0,0,0))
plt.show()
```

```python
#Known as a wireframe plot in python
#Creates a wireframe plot with a viridis color mapping-- viewed at a 75 degree angle 

# Grab some test data.
#X,Y,Z are data values that are created as 2D arrays
X, Y, Z = axes3d.get_test_data(0.3) 

#Normalize to [0,1]-- used for color mapping
Z = (Z-Z.min())/(Z.max()-Z.min())

#Color mapping-- can change function by using other color mapping styles
colors = cm.viridis(Z) #Can customize with various color mapping styles,rcount,ccount,_ = colors.shape

# Plot a basic wireframe.
#rstride = arrayrowstride(step size)
#cstrode = arraycolumnstride(step size)
fig = plt.figure()
ax = fig.gca(projection='3d')

#Setting rotation angle of plot to 45 degrees 
ax.view_init(azim=75)

#rccount= max amount of rows
#ccount = max amount of columns 

surf= ax.plot_surface(X, Y, Z, rcount=100, ccount=100,facecolors=colors, shade=False)

#Setting the tick locators-- choosing tick locations

#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

#Add labels for x,y, and z axes
ax.set_xlabel('X axis', fontsize=10,labelpad=4.2)
ax.set_ylabel('Y axis', fontsize = 10,labelpad=4.2)
ax.set_zlabel('Z axis',fontsize=10,labelpad = 4.2)
ax.set_title('Mesh Plot')
surf.set_facecolor((0,0,0,0))
plt.show()
```
