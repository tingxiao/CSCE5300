## 1. Multiple Line Plot

Letâ€™s make a multiple line plot! The code below creates multiple sets of data from which lines will be created for the plot. This code also creates labels and a legend for the plot.

[This link](https://www.kaggle.com/andyxie/matplotlib-plot-multiple-lines/notebook) describes how to create a plot with multiple lines, and it provides examples of customizing the plot with different line styles, color styles, and more.
<ol type="a">
    <li> Set x-axis and y-axis bounds by using pyplot's <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.ylim.html">ylim()</a> and <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html">xlim()</a> functions. </li>
    <li> At the bottom of [this webpage]( https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html), it shows the different values for colors and markers that can be used in pyplot's <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html"> plot()</a> function. </li>
    <li> [This link]( https://matplotlib.org/gallery/api/two_scales.html) explains how to set different y-axis scales for two lines by using two different axes that share the same x axis. </li>
</ol>

```python
import numpy as np
import matplotlib.pyplot as plt

#Varying y-values and x-values
july_sales = [500,800,700,900,750,600,400]
august_sales = [800,900,600,800,1000,700,650]
days_of_the_week = [1,2,3,4,5,6,7]

#Setting x-labels
week_ticks = ['Week 1','Week 2','Week 3','Week 4','Week 5','Week 6','Week 7']

#Plotting the two different lines
plt.plot(days_of_the_week,july_sales,marker='o',color='b')
plt.plot(days_of_the_week,august_sales,marker='v',color='r')

#Putting a title and labels
plt.title("July Sales vs. August Sales Per Week")
plt.ylabel("Number of Sales")
plt.legend(['July Sales','August Sales'],loc = 'lower center')
plt.xticks(days_of_the_week,week_ticks)
plt.show()
```

## 2. Scatter Plot

Letâ€™s make a scatter plot! The code below creates multiple sets of data from which data points will be set on the figure.

[This link](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) explains the scatter function and its parameters.

[This link](https://jakevdp.github.io/PythonDataScienceHandbook/04.02-simple-scatter-plots.html) also explains how to create a scatter plot with either the plot or scatter function. Feel free to try both ways if you'd like to learn the different techniques of creating a scatter plot.

<ol type="a">
    <li> To change the size of the markers in the scatterplot, you can change the variable 's' which is one of the parameters of the <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html"> scatter() </a> function.</li>
    <li> The following link explains how to create [empty or filled markers]( https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) by changing the value of the 'edgecolors' parameter in the scatter function.</li>
    <li> The following link provides different ways to add a [line of best-fit]( https://stackoverflow.com/questions/19068862/how-to-overplot-a-line-on-a-scatter-plot-in-python/19069001#19069001) to a scatter plot in Python. </li>
</ol>

```python
#Getting x and y values
temperatures = [68,63,74,72,79,78,71,71,69,66,80,58]
customers = [317,355,463,419,507,431,433,434,362,340,540,250]

#Creating the scatter plot
plt.scatter(temperatures, customers, s=100, facecolors ='none', edgecolors='black', alpha=0.5)

#Labelling the plot
plt.xlabel('Average Temperature at the Pool (Degrees Fahrenheit)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers vs. Average Temperature at the Pool')

#Outputting the scatter plot
plt.show()
```

This second example creates a line of best-fit and adds it to the scatter plot.

We create a line of best-fit by using numpy's <a href="https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.unique.html"> unique()</a>,
<a href="https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.poly1d.html"> poly1d() </a>, and <a href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html"> polyfit() </a> functions.

```python
#Getting x and y values
temperatures = [68,63,74,72,79,78,71,71,69,66,80,58]
customers = [317,355,463,419,507,431,433,434,362,340,540,250]


#Creating the scatter plot
plt.scatter(temperatures, customers, s=100, facecolors ='none', edgecolors='black', alpha=0.5)

#Labelling the plot
plt.xlabel('Average Temperature at the Pool (Degrees Fahrenheit)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers vs. Average Temperature at the Pool')
plt.plot(np.unique(temperatures),np.poly1d(np.polyfit(temperatures,customers,1))(np.unique(temperatures)))

#Outputting the scatter plot
plt.show()
```

## 3. Subplot <br>

Letâ€™s make a [subplot]( https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)! The code below creates multiple sets of data from which multiple line plots will be made. It also creates different axes for the line plots and puts them into the same figure. It creates labels for each of the plots.

The first example creates the subplots as line plots for when the data points are ordered; the second example creates the subplots as scatter plots, since the data points are randomly ordered.

The only difference between creating the subplots with lines or scatterplots is the usage of the [plot()]( https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)  function or the [scatter()]( https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) function.

```python
#This example uses the matplotlib subplot() function
#It has ordered data so the data is plotted as lines

#Creating data
#First plot:
average_temp_in_june = [70,70,74,80]
customers_in_june = [237,250,325,457]

#Second plot:
average_temp_in_july = [76,78,80,82]
customers_in_july = [480,489,498,500]

#Third plot
average_temp_in_august = [71,73,80,86]
customers_in_august = [410,500,600,615]

#Creates one plot with random x & y values
#Creating subplot: subplot(nrows, ncols, index)
plt.subplot(2,2,1)
#Plotting
plt.plot(average_temp_in_june,customers_in_june)
plt.xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Creates a second plot with random x & y values
plt.subplot(2,2,2)
plt.plot(average_temp_in_july,customers_in_july)
plt.xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Creates a third plot with random x & y values
plt.subplot(2,1,2)
plt.plot(average_temp_in_august,customers_in_august)
plt.xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Making room for the x-labels (https://stackoverflow.com/questions/6774086/why-is-my-xlabel-cut-off-in-my-matplotlib-plot)
plt.tight_layout()

#Puts all of the plots in the same figure
plt.show()
```

```python
#This example uses the add_subplot function and has randomly-ordered data
#Therefore, it includes scatter plots instead of line plots

#Import package to help style graphs
from matplotlib import style

#Setting the style of the subplot
style.use('fivethirtyeight')
fig = plt.figure()

#Creating data
#First plot:
average_temp_in_june = [70,74,80,70]
customers_in_june = [250,325,457,237]

#Second plot:
average_temp_in_july = [82,76,78,80]
customers_in_july = [500,480,489,498]

#Third plot
average_temp_in_august = [80,71,86,73]
customers_in_august = [600,410,615,500]

#Returns the instance of the Axes.
#E.g., fig.add_subplot(221) divides the figure into two rows (the first 2) and two columns (the second 2) and selected the first of 4 possible locations.
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

#Creates one plot with random x & y values
ax1.set_ylabel('Amount of customers',fontsize=8)
ax1.set_xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
ax1.scatter(average_temp_in_june,customers_in_june)

#Creates a second plot with random x & y values
ax2.set_ylabel('Amount of customers',fontsize=8)
ax2.set_xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
ax2.scatter(average_temp_in_july,customers_in_july)

#Creates a third plot with random x & y values
ax3.set_ylabel('Amount of customers',fontsize=8)
ax3.set_xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
ax3.scatter(average_temp_in_august,customers_in_august)

#Labelling the subplots and creating space
fig.suptitle('Average Temperatures vs. Amount of Customers at the Pool',fontsize=12)
fig.subplots_adjust(hspace=1.0,top=1.0,wspace=1.0)
fig.tight_layout()

#Puts all of the plots in the same figure
plt.show()
```
